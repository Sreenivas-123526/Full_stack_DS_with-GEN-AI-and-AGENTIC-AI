```python
import numpy as np
```

```python
Years=["2020","2021","2022"]
Ydict={"2020":0,"2021":1,"2022":2}

Selected_can_nam=["Sreenu","Pavan","Ravi"]
Sdict={"Sreenu":0,"Pavan":1,"Ravi":2}

Tests_Sreenu=[10,20,30]
Tests_Pavan=[20,30,10]
Tests_Ravi=[1,2,3]
no_of_tests_taken1=np.array([Tests_Sreenu,Tests_Pavan,Tests_Ravi])
```

```python
#LINE CHART

plt.subplot(1,1,1)

plt.plot(Years, no_of_tests_taken1[0], c='green', ls='--', marker='s', ms=8, label=Selected_can_nam[0])
plt.plot(Years, no_of_tests_taken1[1], c='cyan',  ls='--', marker='^', ms=8, label=Selected_can_nam[1])
plt.plot(Years, no_of_tests_taken1[2], c='red',   ls='--', marker='o', ms=8, label=Selected_can_nam[2])

plt.title("Line Chart")
plt.legend()
plt.grid(False)
```

```python
import matplotlib.pyplot as plt
```

```python
#BAR CHART

plt.subplot(1,1,1)

x = np.arange(len(Years))
width = 0.25

plt.bar(x - width, no_of_tests_taken1[0], width, label=Selected_can_nam[0], color='green')
plt.bar(x,no_of_tests_taken1[1], width, label=Selected_can_nam[1], color='cyan')
plt.bar(x + width,  no_of_tests_taken1[2], width, label=Selected_can_nam[2], color='red')
    
plt.xticks(x, Years)
plt.title("Bar Chart")
plt.legend()
plt.grid(True)
```

```python
#PIE CHART

plt.subplot(1,1,1)

# total tests per person
totals = [
    sum(Tests_Sreenu),
    sum(Tests_Pavan),
    sum(Tests_Ravi)
]

plt.pie(totals, labels=Selected_can_nam, autopct='%1.1f%%', colors=['green','cyan','red'])
plt.title("Total Contribution")
```
