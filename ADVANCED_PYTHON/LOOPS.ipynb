```python
#PATTERNS
#RIGHT ANGLED TRIANGLE
for i in range(1,6):
    print('*' *i)
```

```python
#INVERTED RIGHT ANGLED TRIANGLE
for i in range(5,0,-1):
    print('*' *i)
```

```python
#PYRAMID PATTERN
for i in range(1,6):
    print(''*(5-i)+' * '*(2*i-1))
```

```python
#INVERTED RIGHT ANGLE TRIANGLE
for i in range(1,6):
    print(''.join(str(x) for x in range(1,i+1)))

#str(x) for x in range(...)- Converts each number into a string
#''.join(...)-Combines all numbers into a single string (no spaces)
```
