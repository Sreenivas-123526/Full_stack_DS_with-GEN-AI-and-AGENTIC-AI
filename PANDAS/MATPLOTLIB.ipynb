```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(3)


# Call plot() method on the appropriate object
ax[0].plot(x1, np.sin(x1),c= 'r')
ax[1].plot(x1, np.cos(x1), 'ro')
ax[2].plot(x1,np.tanh(x1),'b--')
```

```python
#We create a reference to the figure instance in the fig variable. Then, we ceate a new axis instance axes using the add_axes method in the Figure class instance fig as follows:-

fig = plt.figure()

x2 = np.linspace(0, 5, 10)
y2 = x2 ** 2

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #[left, bottom, width, height]

axes.plot(x2, y2, 'b--')

axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title');
```
